```
people = ['chinnu','Bindhu','chinna','nicy','nikki','cherry','ammu']
age = [18,19,18,20,21,22,18]
weight = [60,66,55,60,59,70,67]
height = [160,135,170,165,173,160,170]
import matplotlib.pyplot as plt
plt.scatter(weight, height)
plt.title("Relationship between Weight and Height of Patients")
plt.ylabel("Height in Meters")
plt.xlabel("Weight of Patients in Kgs")
plt.show()
```

```
plt.figure(figsize=(12,4))
plt.title("People's weight in kgs",fontsize=16,
          fontstyle='italic')
plt.bar(x=people,height=weight, width=0.6,
        color='black',edgecolor='k',alpha=0.6)
plt.xlabel("People",fontsize=10)
plt.ylabel("Weight (in kgs)",fontsize=9)
plt.show()
```

```
import numpy as np
plt.figure(figsize=(6,2))
plt.hist(weight,color='purple',edgecolor='k', alpha=0.75,bins=5)
plt.title("Histogram of patient weight",fontsize=12)
plt.xlabel("Weight in kgs",fontsize=10)
plt.show()
```
